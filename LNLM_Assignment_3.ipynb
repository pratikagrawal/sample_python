```python
%%R
my.OptimizerFunction<-function(Start.Value,Function.To.Optimize,Epsilon,delta=0.001) {
    x_plus_1 <- Start.Value - (Function.To.Optimize(Start.Value)/my.FunctionDeriv(Start.Value,delta))
    if( (x_plus_1 - Start.Value) <= Epsilon) {
        return(Start.Value)
    }
    else {
        return(my.OptimizerFunction(x_plus_1,Function.To.Optimize,Epsilon,(x_plus_1-Start.Value)))
    }
}
```

```python
%%R
my.Function<-function(my.X) {
    return((my.X)^2*3-my.X*5-5)
}
my.FunctionDeriv<-function(my.X,delta=0.001) {
    return((my.Function(my.X+delta) - my.Function(my.X-delta))/(2*delta))
}
```
