#### I am using the Titanic data, and only using the Class and Age as the predictors.  The data set takes a look at the passengers aboard the Titanic, and whether the passenger survived or not.

```python
%%R
data<-read.csv(file="/Users/pagrawal/Documents/Personal/OneDrive/Documents/MS-A/LinearAndNonLinearModels/Assignments/IV/titanic3.csv", header=TRUE , sep=",")
head(subset_data)
```

### Prospective Sampling-

#### I am creating two segments, one is for passengers of age 25 or less, and the other is for passengers of age more than 25

```python
%%R
subset_data=as.data.frame(data[,c(1,5,2)])
subset_data_age_less_than_25=subset(subset_data,age<=25)
subset_data_age_more_than_25=subset(subset_data,age>25)
head(subset_data_less_than_25)
```

```python
%%R
levels(factor(subset_data[,3]))
```

#### Building glm models using the different link functions for both the segments

```python
%%R
modl_1<-glm(survived~.,family=binomial,data=subset_data_age_less_than_25)
modp_1<-glm(survived~.,family=binomial(link=probit),data=subset_data_age_less_than_25)
modc_1<-glm(survived~.,family=binomial(link=cloglog),data=subset_data_age_less_than_25)
 
```

```python
%%R
modl_2<-glm(survived~.,family=binomial,data=subset_data_age_more_than_25)
modp_2<-glm(survived~.,family=binomial(link=probit),data=subset_data_age_more_than_25)
modc_2<-glm(survived~.,family=binomial(link=cloglog),data=subset_data_age_more_than_25)
```

```python
%%R
modl_1$coef
```

#### Plotting the probabilities for first segment (age 25 or less)

```python
%%R
par(mfrow=c(2,1))
x<-seq(-1,10,.2)
pl_1<-ilogit(modl_1$coef[1]+modl_1$coef[2]*x)
pp_1<-pnorm(modp_1$coef[1]+modp_1$coef[2]*x)
pc_1<-1-exp(-exp((modc_1$coef[1]+modc_1$coef[2]*x)))
plot(x,pl_1,type="l",ylab="Probability",xlab="Class")
lines(x,pp_1,lty=2)
lines(x,pc_1,lty=10)
x<-seq(-1,100,.2)
pl_2<-ilogit(modl_1$coef[1]+modl_1$coef[3]*x)
pp_2<-pnorm(modp_1$coef[1]+modp_1$coef[3]*x)
pc_2<-1-exp(-exp((modc_1$coef[1]+modc_1$coef[3]*x)))
plot(x,pl_2,type="l",ylab="Probability",xlab="Age")
lines(x,pp_2,lty=2)
lines(x,pc_2,lty=10)
```

#### From both the graphs above for the first segment we can see that cloglog is a better link function providing a steeper fall-off. For the above segment it can also be seen that both Age and Class are good as far as predictors are concerned, however Class seems to be a better predictor than Age

#### Plotting the probabilities for second segment (age more than 25)

```python
%%R
par(mfrow=c(2,1))
x<-seq(-1,10,.2)
pl_1<-ilogit(modl_2$coef[1]+modl_2$coef[2]*x)
pp_1<-pnorm(modp_2$coef[1]+modp_2$coef[2]*x)
pc_1<-1-exp(-exp((modc_2$coef[1]+modc_2$coef[2]*x)))
plot(x,pl_1,type="l",ylab="Probability",xlab="Class")
lines(x,pp_1,lty=2)
lines(x,pc_1,lty=10)
x<-seq(-10,100,.2)
pl_2<-ilogit(modl_2$coef[1]+modl_2$coef[3]*x)
pp_2<-pnorm(modp_2$coef[1]+modp_2$coef[3]*x)
pc_2<-1-exp(-exp((modc_2$coef[1]+modc_2$coef[3]*x)))
plot(x,pl_2,type="l",ylab="Probability",xlab="Age")
lines(x,pp_2,lty=2)
lines(x,pc_2,lty=10)
```

#### From the graph for Class for the second segment we can see that cloglog is a better link function providing a steeper fall-off. Age is not a good predictor when the segment is for age greater than 25

### Retrospective sampling-

#### Creating the retrospective sample

#### The ratio of not survived to survived is 60:40, and for that purpose I am using a sample of 200 from the original ~1300 passengers with the same ratio

```python
%%R
subset_not_survived=subset(subset_data,survived==0)
subset_survived=subset(subset_data,survived==1)
subset_retrospective_1=subset_not_survived[sample(1:nrow(subset_not_survived), 120, replace = FALSE),]
subset_retrospective_2=subset_survived[sample(1:nrow(subset_survived), 80, replace = FALSE),]
subset_retrospective<-rbind(subset_retrospective_1,subset_retrospective_2)

```

```python
%%R
nrow(subset_retrospective)
```

```python
%%R
modl_1<-glm(survived~.,family=binomial,data=subset_retrospective)
```

#### Plotting the probabilities for Class and Age for model based on retrospective sample 

```python
%%R
par(mfrow=c(2,1))
x<-seq(-1,10,.2)
pl_1<-ilogit(modl_1$coef[1]+modl_1$coef[2]*x)
plot(x,pl_1,type="l",ylab="Probability",xlab="Class")
x<-seq(-1,100,.2)
pl_2<-ilogit(modl_1$coef[1]+modl_1$coef[3]*x)
plot(x,pl_2,type="l",ylab="Probability",xlab="Age")
```

#### In retrospective sampling we cannot use the probit and cloglog link functions. And from above we can see that Age is not a good predictor when using retrospective sampling.
