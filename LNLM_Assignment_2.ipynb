```python
%%R
my.Function<-function(my.X) {
    return((my.X)^2*3-my.X*5-5)
}
```

```python
%%R
my.FunctionDeriv<-function(my.X,delta=0.001) {
    return((my.Function(my.X+delta) - my.Function(my.X-delta))/(2*delta))
}
```

```python
%%R
my.X<-seq(-1,3, by=0.1 )
```

```python
%%R 
plot(my.Function(my.X))
plot(my.FunctionDeriv(my.X,0.001))
```

```python
%%R
my.OptimizerFunction<-function(Start.Value,Function.To.Optimize,Epsilon,delta=0.001) {
    x_plus_1 <- Start.Value - (Function.To.Optimize(Start.Value)/my.FunctionDeriv(Start.Value,delta))
    if( (x_plus_1 - Start.Value) <= Epsilon) {
        return(Start.Value)
    }
    else {
        return(my.OptimizerFunction(x_plus_1,Function.To.Optimize,Epsilon,(x_plus_1-Start.Value)))
    }
}
```

#### Output from custom optimizer function

```python
%%R
my.OptimizerFunction(-10,my.Function,0.001)
```

#### Output from Uniroot function

```python
%%R
uniroot(my.Function,lower=-1,upper=2,extendInt="no", trace=1, tol=0.001)
```

#### Output from Optim function

```python
%%R
optim(-10,
   my.Function,
   method="L-BFGS-B",
   hessian=TRUE,
   lower=c(-Inf,10),
   control=list(trace=1))
```

#### Difference between uniroot and optim

The optim function provides us the point of the function where the maxima or minima has been achieved, and to that effect gives us values for x,y. 
The uniroot function provides the point at which the given equation crosses the x-axis. Even though there could be multiple points of crossing the x-axis, only the first root is printed.
